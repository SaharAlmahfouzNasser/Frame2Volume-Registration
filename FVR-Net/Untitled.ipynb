```python
import nibabel as nib
import numpy as np
import matplotlib.pyplot as plt
import os
```

```python
no = 1030
path = './Result_test/'
```

```python
image_2d = nib.load(path+'us_img_2d_'+str(no)+'.nii.gz').get_fdata()
image_3d = nib.load(path+'us_img_3d_'+str(no)+'.nii.gz').get_fdata()
image_3d_deform = nib.load(path+'us_img_3d_deformable_'+str(no)+'.nii.gz').get_fdata()
image_3d_resampled = nib.load(path+'vol_resampled_'+str(no)+'.nii.gz').get_fdata()
```

```python
fig = plt.figure(30)
plt.subplot(131)
plt.imshow(image_3d[5,:,:],cmap='gray')
plt.axis('off')
plt.title('Preoperative US',fontweight='bold')

plt.subplot(132)
plt.imshow(image_3d_deform[5,:,:],cmap='gray')
plt.axis('off')
plt.title('Intraoperative US',fontweight='bold')

plt.subplot(133)
plt.imshow(image_3d_resampled[5,:,:],cmap='gray')
plt.axis('off')
plt.title('Deformed \n Preoperative US',fontweight='bold')

plt.subplots_adjust(left=0.1,
                    bottom=0.1, 
                    right=0.9, 
                    top=0.9, 
                    wspace=0.4, 
                    hspace=0.4)

plt.savefig("Figure.pdf") 
plt.show()
```
